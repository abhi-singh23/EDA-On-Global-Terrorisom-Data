## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#Extract column which have less nun value.
null_sum=(df.isnull().sum()/len(df))*100
percent_null=null_sum.sort_values(ascending=False)
type(percent_null)
high_null_column=percent_null[percent_null>=50]
less_than_50_null_column=percent_null[~percent_null.isin(high_null_column)]
print(f'Number of column where null value less than fifty percent {len(less_than_50_null_column)}')
less_null_column=less_than_50_null_column.index.to_list()
print(less_null_column)
````

````python
# Write your code to make your dataset analysis ready.
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                   'region_txt':'Region','city':'City','latitude':'Latitude','longitude':'Longitude',
                   'attacktype1_txt':'AttackType','targtype1_txt':'TargetType','targsubtype1_txt':'TargetSubType',
                   'weaptype1_txt':'WeaponType','gname':'Terrorist Group','nkill':'Deaths','nwound':'Wounded','summary':'Summary','motive':'Motive',},inplace=True)

#Extracting Necessary Columns
df=df[['eventid','Year','Month','Day','Country','State','Region','City','Latitude','Longitude','AttackType','TargetType','TargetSubType','Deaths','suicide','success','Wounded','Summary'
,'Terrorist Group','WeaponType','Motive',]]
````
